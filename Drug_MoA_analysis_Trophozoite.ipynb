```python
data_list=[]
data_list.append(df_T)
data_list.append(df_V)
df_sum=pd.concat(data_list, axis=1)

genes_zero_values = df_sum.index[df_sum.values.sum(axis=1) == 0]
df_sum.drop(genes_zero_values, inplace=True)
print('After filtering:', len(df_sum), 'counts')

Total_counts = df_sum.values.sum(axis=0)
n_genes = (df_sum.values > 0.0).astype(int).sum(axis=0)

sample_labels = df_sum.columns
samples_to_remove = [sample_label for sample_label, n_gene in zip(sample_labels, n_genes) if n_gene <= 500]
df_sum.drop(columns=samples_to_remove, inplace=True)

print(n_genes)
print('Stats for counts: max=', np.max(Total_counts),
      'min=', np.min(Total_counts),
      f'average= {np.average(Total_counts):.2f}')

print('Stats for detected genes: max=', np.max(n_genes),
     'min=', np.min(n_genes),
     f'average={np.average(n_genes):.2f}')

display(df_sum)
```

```python
normalized = 10000 * df_sum.values / df_sum.values.sum(axis=0)

lognormalized = np.log1p(normalized)
df_lognormalized = pd.DataFrame(lognormalized, index=df_sum.index, columns=df_sum.columns)

mean_expressions = np.expm1(df_lognormalized.values).mean(axis=1)
mean_expressions[mean_expressions == 0] = 1e-12
log_mean_expressions = np.log1p(mean_expressions)
genes_stats = pd.DataFrame(log_mean_expressions, index=df_lognormalized.index, columns=['AverageExpr'])

dispersion = np.expm1(df_lognormalized.values).var(axis=1, ddof=1)/ mean_expressions
dispersion[dispersion == 0] = np.nan

dispersion = np.log(dispersion)
genes_stats['Dispersion'] = dispersion

display(genes_stats)
```

```python
n_highly_variable = 1000
genes_disp = genes_stats.sort_values(by=['Dispersion'], ascending=False)
highly_variable_genes = genes_disp.head(n_highly_variable).index

isHVG = genes_stats.index.isin(highly_variable_genes)

fig, ax = plt.subplots()

ax.scatter(genes_stats.loc[~isHVG, 'AverageExpr'],
          genes_stats.loc[~isHVG, 'Dispersion'],
          c='gray', s=6, alpha=0.5, label = 'non-variable')

ax.scatter(genes_stats.loc[isHVG, 'AverageExpr'],
          genes_stats.loc[isHVG, 'Dispersion'],
          c='red', s=6, alpha=0.5, label = '{} highly variable'.format(n_highly_variable))

ax.set_xlabel('Mean expresson of genes')
ax.set_ylabel('Dispersion of genes')
plt.legend()

plt.show()


df_HVGs = df_lognormalized.loc[isHVG, :]
print(df_HVGs.shape)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_values = scaler.fit_transform(df_HVGs.values.T)
df_scaled = pd.DataFrame(scaled_values.T, index=df_HVGs.index, columns=df_HVGs.columns)
display(df_scaled)


import anndata as ad
import scanpy as sc


adata = ad.AnnData(df_scaled.T)
sc.tl.pca(adata, svd_solver='arpack')
sc.pl.pca_variance_ratio(adata, log=True)
```
